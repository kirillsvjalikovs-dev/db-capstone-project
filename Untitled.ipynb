```python
show_tables_query = "SHOW TABLES"
cursor.execute(show_tables_query)

results = cursor.fetchall()

print(results)
```

```python
cursor.execute("""
CREATE TABLE IF NOT EXISTS Customers (
    CustomerID INT PRIMARY KEY,
    FullName VARCHAR(100),
    ContactNumber VARCHAR(20)
)
""")
```

```python
cursor.execute("""
INSERT INTO Customers (CustomerID, FullName, ContactNumber)
VALUES
(1, 'John Doe', '12345678'),
(2, 'Jane Smith', '87654321')
""")

connection.commit()
```

```python
join_query = """
SELECT 
    Customers.FullName,
    Customers.ContactNumber,
    Orders.total_cost
FROM Customers
JOIN Orders
ON Customers.CustomerID = Orders.customer_id
WHERE Orders.total_cost > 60
"""
cursor.execute(join_query)

results = cursor.fetchall()

for row in results:
    print(row)
```
